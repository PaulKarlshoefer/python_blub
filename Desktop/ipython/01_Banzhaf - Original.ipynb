## Power set

```python
#itertools.combination return an iterator over all combinations
combinations_2 = itertools.combinations((1,2,3,4), 2 )
combinations_3 = itertools.combinations((1,2,3,4), 3 )
print combinations_2

for c in combinations_2:
    print c
    
for c in combinations_3:
    print c
```

All combinations have to be put together to a single *iterator* which is generated by `from_iterable`

```python
combinations_chain = itertools.chain.from_iterable( itertools.combinations((1,2,3,4),i) for i in range(1,5))

for c in combinations_chain:
    print c
```

Now we define a function which creates a power set. Note that we create a power set which is containing the empty set as well.

```python
def power_set(X):
    "Iterator over all partial sets of X (return tupels)"
    return itertools.chain.from_iterable( itertools.combinations(X, i) for i in range(len(X)+1) )
```

## Computation of the Banzhaf indices

```python
def sum_of_weights(w,S):
    "sum of weights of a coalition"
    sum = 0
    for i in S:
        sum = sum + w[i]
    return sum
```

```python
def banzhaf(w, q):
    """determine Banzhaf indices and print a table

    w: Dictionary with parties as key and weights as data
    q: Quotum
    """  
  
    S = sorted(w.keys())
    
    # create dictionary with keys of S and values 0
    b = dict([j, 0] for j in S)
    
    # Heading
    for j in S:
        print "%5s" % j,
    print "\n" + "------"*len(S)
    for j in S:
        print "%5d" % w[j],
    print "votes "
    
    # loop over all coalitions
    for T in power_set(S):
        summe_T = sum_of_weights(w,T)
        # loop over all parties
        for j in S:
            # count and print table entry
            if j in T:
                print "    *",
            # only if summe_T is smaller then q, the vote has influence
            elif (q-summe_T<= w[j]) and (summe_T<q): # etwas laenger aber besser lesbar
                print "%5d" % (summe_T + w[j]),
                # j can decide the vote -> counted
                b[j] = b[j] + 1
            else:
                print "     ",
        print "(%d)" % summe_T
        
    # total banzhaf index
    print "------"*len(S)
    for j in S:
        print "%5d" % b[j],
    print "Total Banzhaf-Index"
    
    # relative Banzhaf-Index
    print "------"*len(S)
    b_sum = float(sum(b.values()))
    for j in S:
        print "%5.2f" % (b[j]/b_sum),
    print "relative Banzhaf-Index"

    # heading, again
    for j in S:
        print "%5s" % j,
    print
    for j in S:
        print "%5d" % w[j],
    print "votes "
    
```
