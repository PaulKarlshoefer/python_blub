# Allgemeiner Scheduler

```python
scheduling=nx.DiGraph()

scheduling.add_edges_from([('a1','a3'),('a2','a3'),('a2','a4'),('a3','a4'),('a1','a4'),('a4','a5'),('a2','a5'),('a3','a6')])
```

```python
def draw_graph(g,pos=None):
    figure(figsize=(16,10))
    subplot(111)
    if pos==None:
        pos=nx.spring_layout(g)
    nx.draw_networkx(g,pos)
    r = 0.05
    shift=0.02
    options={'bbox':dict(facecolor='white')}
    nodes=g.node
    for node in g.nodes():
        try:
            text(pos[node][0]-r-shift,pos[node][1]+r-shift,nodes[node]['s'],**options)
            text(pos[node][0]+r-shift,pos[node][1]+r-shift,nodes[node]['c'],**options)
        except (StandardError):
            pass
        
        try:
            text(pos[node][0]-r-shift,pos[node][1]-r-shift,nodes[node]['S'],**options)
            text(pos[node][0]+r-shift,pos[node][1]-r-shift,nodes[node]['C'],**options)
        except (StandardError):
            pass
    ax=gca()
    ax.set_aspect('equal')
    ax.set_axis_off()
    return pos

pos=draw_graph(scheduling)
```

```python
def add_start_end(g):
    '''adding a start and an end node'''
    g.add_node('start')
    g.add_node('end')
    
    g.node['start']['t']=0
    g.node['end']['t']=0
    
    ##loop for adding a starting and an end node
    for node in g.nodes_iter():
        if node!='start' and node!='end':
            pred = g.predecessors(node)
            succ = g.successors(node)
            if len(pred)==0:
                g.add_edge('start',node)
            if len(succ)==0:
                g.add_edge(node,'end')
            
add_start_end(scheduling)
draw_graph(scheduling)
help(add_start_end)
```

```python
nodes=scheduling.node

nodes['a1']['t']=10
nodes['a2']['t']=40
nodes['a3']['t']=30
nodes['a4']['t']=10
nodes['a5']['t']=10
nodes['a6']['t']=60

def init_nodes(g):
    '''initialization '''
    for node in g.nodes():
        g.node[node]['c']=0
        g.node[node]['s']=0
        g.node[node]['C']=float('Inf')
        g.node[node]['S']=0
    

init_nodes(scheduling)
pos = draw_graph(scheduling)
```

```python
##topological sorting of all of the nodes
top = nx.algorithms.topological_sort(scheduling)
print top
```

```python
def earliest_starting(g):
    '''calculation of the earliest starting times'''
    top = nx.algorithms.topological_sort(g)
    for node in top[1:]:
        for in_node,buf in g.in_edges(node):
            if g.node[in_node]['c'] > g.node[node]['s']:
                g.node[node]['s']=g.node[in_node]['c']
        g.node[node]['c']=g.node[node]['s']+g.node[node]['t']
        
    g.node['end']['S'] = g.node['end']['s']
    g.node['end']['C'] = g.node['end']['s']

earliest_starting(scheduling)

pos = draw_graph(scheduling,pos)
```

```python
def latest_starting(g):
    '''calculation of the latest starting times'''
    top = nx.algorithms.topological_sort(g)
    nodes=g.node
    for node in top[-2::-1]:
        for buf,out_node in g.out_edges(node):
            if nodes[out_node]['S'] < nodes[node]['C']:
                nodes[node]['C'] = nodes[out_node]['S']
        nodes[node]['S'] = nodes[node]['C'] - nodes[node]['t']
        
latest_starting(scheduling)
    
pos  = draw_graph(scheduling,pos)
```

```python
def update_node_pos(g):
    '''update of the nodes positions'''
    top = nx.algorithms.topological_sort(g)
    pos = {}
    l=[]
    pos['start']=(0,0.5)
    nodes=g.node
    
    for node in top[1:]:
        l.append(len(nx.shortest_path(g,source='start',target=node))-1)
        pos[node]=(l[-1],0.)
        
    for node in top[1:]:
        pos[node]=((float(nodes[node]['c'])+nodes[node]['s'])/(nodes['end']['s']),rand())
    return pos

pos = update_node_pos(scheduling)

pos = draw_graph(scheduling,pos)
```

```python
def all_graph_operations(g):
    add_start_end(g)
    init_nodes(g)
    earliest_starting(g)
    latest_starting(g)
    pos = update_node_pos(g)
    return pos
```

# Job Shop

```python
js = nx.DiGraph()
js.add_edges_from([('a11','a12'),('a12','a13'),('a13','a14'),('a14','a15')])
js.add_edges_from([('a21','a22')])

for n in ['a11','a12','a13','a14','a15']:
    js.node[n]['t']=2
    
for n in ['a21','a22']:
    js.node[n]['t']=3

pos = draw_graph(js)
```

```python
pos = all_graph_operations(js)

draw_graph(js,pos)
```

```python
js.add_edge('a13','a22')
pos = all_graph_operations(js)

draw_graph(js,pos)
js.remove_edge('a13','a22')
```
