```python
import math
from numpy import random as random
from matplotlib import pylab as pl

# make nice plots
import seaborn as sns
sns.set_context('notebook', font_scale=1.2)
sns.set_style('whitegrid')
%pylab inline
```

Helferfunktion fuer die Zufallsvariable:

```python
def T(t0, tG, p):
    """ Zeit fuer einen Hamburger: mit WS p ist T=t0+tG, sonst t0
    """
    if random.random() < p:
        return t0+tG
    else:
        return t0
```

Die eigentlichen Simulationsfunktionen

```python
def simulation(t0, tG, p, N, M, fig):
    """ N Hamburger braten, waehrenddessen M Beobachter vorbeischicken
    t0: typische Bratzeit
    tG: Zusatzzeit beim Gurkenoeffnen
    p: Wahrscheinlich dass eine Gurkendose geoeffnet werden soll
    N: Burgeranzahl
    M: Beobachteranzahl
    """
    ET = t0 + tG*p    # Vergleichswerte fuer die Ausgabe
    sigmaT = math.sqrt(p*(1.-p))*tG
    rhoT = sigmaT/ET
    EVRZ = ET*(1+rhoT**2)/2.
    print 'T: Erwartungswert %.3g, Standardabweichung %.2g, Variationskoeff. %.3g'\
          % (ET, sigmaT, rhoT)
    print 'Erwartete Vorwaertsrekurrenzzeit: %.3g' % EVRZ
    
    Ti = [T(t0, tG, p) for i in range(N)] # Bestimme jeweilige Zeitdauer fuer alle N Burger
    Tsum = [sum(Ti[0:i]) for i in range(1,N+1)] # Tsum[j] = Summe Ti[0]..Ti[j]
    tend = Tsum[-1] # Dauer des ganzen Versuchs
    print 'Mittelwert der Ti: %.3g' % (tend/N) # mittlere Burgerbratzeit
    TB = []    # Eintreffen der Beobachter
    VRZ = []   # Zugehoerige Vorwaertsrekurrenzzeit
    for j in range(M):
        tb = random.random()*tend # Beobachter trifft ein...
        for t in Tsum:
            if t>=tb:             # ... sieht den naechsten Hamburger...
                break
        TB.append(tb)
        VRZ.append(t-tb)          # ... und notiert die Zeit
    print 'Mittelwert der VRZ: %.3g' % (sum(VRZ)/M)

    if fig != None:
        pl.figure(num=fig.number)
        pl.xlabel('$t$')
        pl.ylabel('VRZ')
        ecken = [[0., 0.]]
        t = 0.
        for ti in Ti:
            ecken.append([t, ti])
            ecken.append([t + ti,0.])
            t = t + ti
        ecken = np.array(ecken)
        pl.plot(ecken[:,0], ecken[:,1], 'k-', label='Wartezeiten', alpha=0.5)
        pkt = []
        for i in range(M):
            pkt.append([TB[i], VRZ[i]])
        pkt = np.array(pkt)
        pl.plot(pkt[:,0], pkt[:,1], 'db', label='Beobachtungen')
        pl.legend(loc='upper right', bbox_to_anchor=(1.5, 1.0))
    return fig


def simulationBild(t0, tG, p, N, M):
    ''' Simulieren und Bilder generieren
    t0: typische Bratzeit
    tG: Zusatzzeit beim Gurkenoeffnen
    p: Wahrscheinlich dass eine Gurkendose geoeffnet werden soll
    N: Burgeranzahl
    M: Beobachteranzahl
    '''
    fig = pl.figure()
    simulation(t0, tG, p, N, M, fig)
    return fig
```
