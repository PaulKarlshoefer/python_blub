```python
%pylab inline
# get the animation widget from  https://github.com/jakevdp/JSAnimation.git
from JSAnimation.IPython_display import display_animation
import numpy as np
import matplotlib.animation as animation
import matplotlib.pyplot as py
```

```python
class fuzzy_set:
	def __init__(self,names,points):
		'''
		constructor:
		requirese the linguistic variables and its respective values
		'''
		self.names=names
		self.points=points
		
	def get_mu(self,name,x):
		'''
		returns the truth value of a certain fuzzy-node
		'''
		p=self.points
		i=self.names.index(name)
		if i==0:
			return self.lin1Dr(p[0],p[1],x)
		elif i==len(self.names)-1:
			return self.lin1Dl(p[-2],p[-1],x)
		else:
			if x<self.points[i]:
				return self.lin1Dl(p[i-1],p[i],x)
			else:
				return self.lin1Dr(p[i],p[i+1],x)

            
	def lin1Dl(self,l,m,x):
		'''
		left part of hat-function
		'''
		if x<l:
			return 0
		elif x<m:
			return 1./(m-l)*x+1-m/(m-l)
		else:
			return 1.

        
	def lin1Dr(self,m,r,x):
		'''
		right part of hat-function
		'''
		if x<m:
			return 1.0
		elif x<r:
			return x/(m-r)+1-m/(m-r)
		else:
			return 0.
```

```python
class fuzzy_control:
	def __init__(self,p1,p2,impl,dz=1.):
		self.p1=p1
		self.p2=p2
		self.impl=impl
		self.rules=[]
		self.dz=dz
		
	def add_rule(self,p1_ling,p2_ling,impl_ling):
		self.rules.append([p1_ling,p2_ling,impl_ling])
    
        
	def fuzzify(self,fuzzy_set,ling,x):
		return fuzzy_set.get_mu(ling,x)
    
    
	def aggregate(self,ling1,ling2,x,y):
		return min(self.fuzzify(self.p1,ling1,x),
                   self.fuzzify(self.p2,ling2,y))
	
    
	def implication(self,ling1,ling2,ling_impl,x,y,z):
		return min(self.fuzzify(self.impl,ling_impl,z),
                   self.aggregate(ling1,ling2,x,y))
	
    
	def accumulate_rules(self,x,y,z):
		accu=0.0
		for rule in self.rules:
			b=self.implication(rule[0],rule[1],rule[2],x,y,z)
			if b>accu:
				accu=b
		return accu
	
    
	def defuzzify(self,x,y):
		'''find center of mass'''
		range_z=np.arange(self.impl.points[0],self.impl.points[-1],self.dz)
		M0=0.0
		M1=0.0
		for z in range_z:
			rho = self.accumulate_rules(x,y,z)
			M0 += rho
			M1 += rho*z
		return M1/M0
	
    
	def get_accumulated_set(self,x,y):
		range_z=np.arange(self.impl.points[0],self.impl.points[-1],self.dz)
		rho=[]
		for z in range_z:
			rho.append(self.accumulate_rules(x,y,z))
		return np.array(rho)
	
    
	def defuzzify_wrapper(self,x):
		return self.defuzzify(x[2],x[3])
```

```python
def create_animation_object(s,phi,L):
	fig=py.figure()
	ax1=fig.add_subplot(111)
	ax1.set_aspect('equal')
	ax1.set_xlim((-L,L))
	ax1.set_ylim((0,L))
	line,=ax1.plot([s[0],s[0]+L*np.sin(phi[0])],[0,L*np.cos(phi[0])],'r',lw=3,animated=True)
    
	def animate(i):
		ax1.set_xlim((-L,L)+s[i])
		line.set_xdata([s[i],s[i]+L*np.sin(phi[i])])
		line.set_ydata([0,L*np.cos(phi[i])])
		py.draw()
		return line,
  
	def init():
		line.set_ydata(np.ma.array(s, mask=True))
		line.set_xdata(np.ma.array(s, mask=True))
		return line,
    
	ani = animation.FuncAnimation(fig,animate,range(len(s)),interval=10,blit=True,init_func=init)
	return ani


def simulation(A,B,x0,u,L,n=10):
	erg=[]
	x=x0.copy()
	for i in range(n):
		x_new = euler_step(A,B,x,u)
		erg.append(x_new)
		x=x_new
	erg=np.array(erg)
	return create_animation_object(erg[:,0],erg[:,2],L)


def euler_step(A,B, x, u,dt=0.01):
	return x+dt*(np.dot(A,x)+B*u(x))


def u_simple(x):
	return 0.0
```

```python
##some constants
M=0.981
m=0.08
L=0.312

##linearised inverse pendulum
def getAB(M,m,L,g=9.81):
	x1=-3.*m*g/(m+4*M)
	x2=3.*g*(m+M)/(L*(m+4*M))
	x3=-4./(m*4*M)
	x4=3./(L*(m+4*M))
	return np.array([[0,1,0,0],[0,0,x1,0],[0,0,0,1],[0,0,x2,0]]),np.array([0,x3,0,x4])


##linguistic variables
ling=['NL','N','Z','P','PL']

##create fuzzy nodes
winkel=np.array([-10,-5,0,5,10])/360.*2.*np.pi

##create fuzzy sets
sphi=fuzzy_set(ling,winkel)
sphip=fuzzy_set(ling,winkel*4.)
su=fuzzy_set(ling,np.linspace(-15.,15,5))

##get system
A,B=getAB(M,m,L)

##create rule
rule_basis=\
[[  0,   0,'PL', 'P', 'Z'],\
[   0,'PL', 'P', 'Z', 'N'],\
['PL', 'P', 'Z', 'N','NL'],\
[ 'P', 'Z', 'N','NL',   0],\
[ 'Z', 'N','NL',   0,   0]]

#rule_basis=\
#[[  0,   0,'PL',   0,   0],\
#[   0,'PL', 'P', 'Z',   0],\
#['PL', 'P', 'Z', 'N','NL'],\
#[   0, 'Z', 'N','NL',   0],\
#[   0,   0,'NL',   0,   0]]


##create controller
fuzzy_controller = fuzzy_control(sphi,sphip,su)

##feed rules into controller
for i in range(len(rule_basis)):
	for j in  range(len(rule_basis[0])):
		if rule_basis[i][j]!=0:
			fuzzy_controller.add_rule(fuzzy_controller.p1.names[i],\
			fuzzy_controller.p2.names[j],rule_basis[i][j])

##initial condition
x0=np.array([0.,0.,0.8,0.0])


anim = simulation(A,B,x0,fuzzy_controller.defuzzify_wrapper,L,n=200)
display_animation(anim, default_mode='loop')
```
