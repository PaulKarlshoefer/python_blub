## profit function

Profit of a farmer $i$ from strawberry farming is a difference between $$K_i=x_1^2$$ and the price per unit $$p=120-2\cdot(x_1+x_2)$$.

```python
preis_ = 120 - 2*(x1_+x2_)
preis_
```

### farmer 1:

```python
K1_ = x1_**2
K1_
```

```python
GewinnBauer1_ = preis_*x1_-K1_
GewinnBauer1_
```

Here we transfer the symbolic function into a regular python function which can be easily evaluated.

```python
#def GewinnBauer1(x1,x2):
#    return (120-2*(x1+x2))*x1-x1**2

GewinnBauer1 = sym.lambdify( (x1_,x2_) , GewinnBauer1_ )
```

Now we plot this function:

```python
ax=subplot(111,projection='3d')            #init 3D
Z = GewinnBauer1(X1,X2).clip(0,1500)       # result is cropped to [0,1500] (no negative profit)
ax.plot_surface(X1,X2,Z,**outputOptions)   # plot of the profit function
ax.view_init(30,45)                        # view angle
```

### Farmer 2:

```python
K2_ = x2_**2
K2_
```

```python
GewinnBauer2_ = preis_*x2_-K2_
GewinnBauer2_
```

```python
def GewinnBauer2(x1,x2):
    return (120-2*(x1+x2))*x2-x2**2
```

```python
ax=subplot(111,projection='3d')
Z = GewinnBauer2(X1,X2)
ax.plot_surface(X1,X2,Z.clip(0,1500),**outputOptions)
ax.view_init(30,45)
```

## Reaction function $r_1$ and $r_2$

optimal amount of strawberries of farmer 1 for a given $x_2$

```python
GewinnBauer1_diff_ = sym.diff(GewinnBauer1_,x1_)
GewinnBauer1_diff_
```

```python
r1_ = sym.solve(GewinnBauer1_diff_,x1_)[0]
r1=sym.lambdify(x2_,r1_)
r1_
```

same for farmer 2:

```python
r2_ = sym.solve(sym.diff(GewinnBauer2_,x2_),x2_)[0]
r2=sym.lambdify(x1_,r2_)
r2_
```

the function describe the profits along the line $(x_1,x_2)=(r_1(t),t)$

```python
ax=subplot(111,projection='3d')
t=arange(0,60,0.1)                                   # numpy array with all values between [0,60] 
ax.plot(r1(t),t,GewinnBauer1(r1(t),t))               # plot of the line
ax.plot(r1(t),t,GewinnBauer2(r1(t),t).clip(0,1500))
ax.set_xlim((0,60))                                  # set the axis for consitency
ax.view_init(30,45)
```

Same for farmer 2:

```python
ax=subplot(111,projection='3d')
t=arange(0,60,0.1)

ax.plot(r1(t),t,GewinnBauer1(r1(t),t),'k')
ax.plot(r1(t),t,GewinnBauer2(r1(t),t).clip(0,1500),'r')

ax.plot(t,r2(t),GewinnBauer1(t,r2(t)).clip(0,1500),'k')
ax.plot(t,r2(t),GewinnBauer2(t,r2(t)),'r')

ax.view_init(30,45)
```

## Equilibrium

### iterative solution

```python
x1 = x2 = 30
trajectory = []
change = 1.
i = 0
while (change>0.0001):
    i+=1
    change = abs(x2 - r2(x1))
    x2 = r2(x1)
    trajectory.append([x1,x2])
    x1 = r1(x2)
    
trajectory = array(trajectory)
print i,' iterations to find equilibrium at: ',trajectory[-1]

plot(trajectory[:,0],trajectory[:,1],'o-')
```

### direct solution

solve the linear system

```python
x_GG_=sym.solve(r1_.subs(x2_,r2_)-x1_,x1_)[0]
x_GG_
```

```python
print GewinnBauer1(x_GG_,x_GG_),GewinnBauer2(x_GG_,x_GG_)
```

## The Strawberry Cartel

what could they make in profit if they are cooperating?

```python
def Gesamtgewinn(x1,x2):
    return GewinnBauer1(x1,x2) + GewinnBauer2(x1,x2)
print Gesamtgewinn(x_GG_,x_GG_)
```

```python
Gesamtgewinn_=Gesamtgewinn(x1_,x2_)
Gesamtgewinn_
```

```python
ax = subplot(111,projection='3d')
Z=Gesamtgewinn(X1,X2).clip(0,1500)
ax.plot_surface(X1,X2,Z,**outputOptions)
ax.view_init(30,120)
```
