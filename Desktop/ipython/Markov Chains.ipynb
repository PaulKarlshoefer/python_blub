```python
import numpy as np
import scipy as sp
from numpy import random as random
from matplotlib import pylab as pl

# make nice plots
import seaborn as sns
sns.set_context('notebook', font_scale=1.2)
sns.set_style('whitegrid')
%pylab inline
```

```python
# these functions remain the same for both excercises

def X_exponential(theta):
    """ negative exponentially distributed random variable (parameter theta)
    """
    r = random.random()
    return - np.log(1. - r)/theta # logarithmic transform


def newestEvent(lambd, mu):
    """ Event  ('arrival', t) oder ('service', t)
    
    Arrival rate lambda, service rate mu
    mu<=0: next event needs to be an arrival (for empty waiting queue)
    """
    if mu <= 0.:
        return('arrival', X_exponential(lambd))
    t_arrival = X_exponential(lambd)
    t_service = X_exponential(mu)
    # if arrival event would occur after service: return service event
    if t_arrival > t_service: 
        return('service', t_service)
    else:
        return('arrival', t_arrival)


def filling(s, fig=None):
    """Diagram of filling over the time for simulation event s
    
    If fig is set: draw the diagram
    """
    pkt = [s[0]]
    f_old = s[0][1]
    for i in range(1, len(s)):
        si = s[i]
        pkt.append([si[0], f_old]) # Draw old fillint until the current time t...
        pkt.append(si)             # ... and then the new filling
        f_old = si[1]
    if fig != None:
        pkt = np.array(pkt)
        pl.figure(num=fig.number)
        pl.title('Filling')
        pl.xlabel('$t$')
        pl.ylabel('Filling')
        pl.plot(pkt[:, 0], pkt[:, 1], '-')
    return pkt
    

def cumulativeFilling(s, fig=None):
    """Cumulative filling for simulation result s
    
    Integral over filling between t=0 and t=simulation end
    If fig is set: plot integral over time
    """
    sum = 0.
    t_old = 0.
    f_old = s[0][1]
    pkt = [[t_old, sum]]
    for i in range(1, len(s)):
        si = s[i]
        sum = sum + (si[0]-t_old)*f_old # from the last event until now:
                                        # increase according to filling until now
        t_old, f_old = si
        pkt.append([t_old, sum])
    if fig != None:
        pkt = np.array(pkt)
        pl.figure(num=fig.number)
        pl.title('Cumulative filling')
        pl.xlabel('$t$')
        pl.ylabel('Cumulative filling')
        pl.plot(pkt[:, 0], pkt[:,1], '-')
    return sum
        

def averageFilling(s, fig=None):
    """avarage filling over time for simulation result s
    
    Cumulative filing over the simulation duration
    If fig is set: plot a diagram over time
    """
    sum = 0.
    t_old = 0.
    f_old = s[0][1]
    pkt = [[t_old, sum]]
    for i in range(1, len(s)):
        si = s[i]
        sum = sum + (si[0]-t_old)*f_old
        t_old, f_old = si
        pkt.append([t_old, sum/t_old])
    if pl != None:
        pkt = np.array(pkt)
        pl.figure(num=fig.number)
        fig.set_figwidth(10)
        fig.set_figheight(5)
        pl.subplot(121)
        pl.title('Avarage filling')
        pl.xlabel('$t$')
        pl.ylabel('avarage filling')
        pl.plot(pkt[:, 0], pkt[:, 1], '.')
        pl.subplot(122)
        pl.hist(pkt[:, 1], normed=True, bins=np.max(pkt[:, 1]), align='left')
        pl.xlim((-0.5, np.max(pkt[:, 1])+0.5))
        pl.xlabel('Filling')
        pl.ylabel('Normed frequency')
    return sum/t_old
    
    
```

## 1 Simple Post Office: Classical $ M | M | 1$

```python
def  dweillingTimes(s, lambd, mu, fig = None):
    """Prints a table with dwilling times for the simulation result s
    
    Rates lambda and mu for evaluated comparison values
    If fig is set: plot the empirical and expected dweilling time histograms
    """
    p = dict()   # p for a key i the dwelling time in the state i
    t_old = 0.         # old stated starts
    f_old = s[0][1]    # and has this filling
    f_max = f_old     
    for i in range(1, len(s)):
        si = s[i]
        p[f_old] = p.get(f_old, 0.) + si[0] - t_old # and to the old state
        if f_old > f_max:
            f_max = f_old
        t_old, f_old = si                           
    t_end = s[-1][0]
    dwell_empir = np.zeros(f_max+1) # empirical dwelling times
    if lambd < mu:
        dwell_expect = np.zeros(f_max+1) # expected dwelling times
    for i in range(0, f_max+1):
        ti = p.get(i, 0.0)       # dwelling time
        dwell_empir[i] = ti 
        if fig == None: # print instead of plot
            ti_proz = ti*100./t_end
            print 'Dwelling time in state %i: %.3g (%.3g%%)'\
                  % (i, ti, ti_proz),
        if lambd<mu:
            p_i = (1-lambd/mu)*(lambd/mu)**i # value for comparison
            dwell_expect[i] = p_i * t_end
            if fig == None: # print instead of plot
                Eti_proz = 100.*p_i      
                Eti = p_i*t_end
                print ', expected:  %.3g (%.3g%%)'\
                      % (Eti, Eti_proz)
                    
    if fig != None:
        pl.figure(num=fig.number)
        pl.plot(dwell_empir, '-r', label="Empirical dwelling time")
        if lambd < mu:
            pl.plot(dwell_expect, '-b', label="Expected dwelling time")
        pl.xlabel("State")
        pl.ylabel("Dwelling time")
        pl.title("Dwelling time in states")
        pl.legend(loc='best')
     
   
def simulation_driver(lambd, mu, f, n):
    """ Pictures for M|M|1: rates lambda und mu, initial filling f, n steps
    """
    # make sure the rates are floats
    lambd = float(lambd)
    mu = float(mu)
    
    s = simulate_simple_post(lambd, mu, f, n)
    fig1 = pl.figure()
    #fig2 = pl.figure()
    fig3 = pl.figure()
    fig4 = pl.figure()
    filling(s, fig1)
    #cumulativeFilling(s, fig2)
    mf = averageFilling(s, fig3)
    print "workload %.3f" %(lambd/mu)
    print 'average filling: %.3g' % mf
    if lambd < mu:
        r = lambd/mu
        print 'expected filling: %.3g' % (r/(1.-r))
    dweillingTimes(s, lambd, mu, fig4)
    

def simulate_simple_post(lambd, mu, f, n):
    """ n simulation step of M|M|1, rates lambda, mu, initial filling f
    
    events: a list of n+1 tuples (t, f)
    Fillings at the beginning and after n events
    """
    print "Simulate Classical $M | M | 1$"
    t = 0.
    evnt = [(t, f)]
    for i in range(n):
        if f>0:
            typ, dt = newestEvent(lambd, mu)
        else:
            typ, dt = newestEvent(lambd, -1.)
        t = t + dt
        if typ == 'arrival':
            f = f + 1
        else:
            f = f - 1
        evnt.append((t, f))
    return evnt
```

```python
# small run
simulation_driver(8., 12., 10, 1000)
```

```python
# large run
simulation_driver(8., 12., 10, 10000)
```

```python
# overload
simulation_driver(8., 4., 0, 10000)
```

## 2 A Second Post Office: Modification of $ M | M | 1$

```python
def  dweillingTimes(s, lambd, mu, fig = None):
    """Prints a table with dwilling times for the simulation result s
    
    Rates lambda and mu for evaluated comparison values
    If fig is set: plot the empirical and expected dweilling time histograms
    """
    p = dict()   # p for a key i the dwelling time in the state i
    t_old = 0.         # old stated starts
    f_old = s[0][1]    # and has this filling
    f_max = f_old     
    for i in range(1, len(s)):
        si = s[i]
        p[f_old] = p.get(f_old, 0.) + si[0] - t_old # and to the old state
        if f_old > f_max:
            f_max = f_old
        t_old, f_old = si                           
    t_end = s[-1][0]
    dwell_empir = np.zeros(f_max+1) # empirical dwelling times
    if lambd < mu:
        dwell_expect = np.zeros(f_max+1) # expected dwelling times

    for i in range(0, f_max+1):
        ti = p.get(i, 0.0)       # dwelling time
        dwell_empir[i] = ti 
        if fig == None: # print instead of plot
            ti_proz = ti*100./t_end
            print 'Dwelling time in state %i: %.3g (%.3g%%)'\
                  % (i, ti, ti_proz),
        if lambd<mu:
            # MODIFIED STATE PROBABILITY
            p_i = np.exp(-lambd/mu) if i == 0 else p_i * (lambd/mu)/i
            dwell_expect[i] = p_i * t_end
            if fig == None: # print instead of plot
                Eti_proz = 100.*p_i      
                Eti = p_i*t_end
                print ', expected:  %.3g (%.3g%%)'\
                      % (Eti, Eti_proz)
                    
    if fig != None:
        pl.figure(num=fig.number)
        pl.plot(dwell_empir, '-r', label="Empirical dwelling time")
        if lambd < mu:
            pl.plot(dwell_expect, '-b', label="Expected dwelling time")
        pl.xlabel("State")
        pl.ylabel("Dwelling time")
        pl.title("Dwelling time in states")
        pl.legend(loc='best')

        
def simulate_modified_postoffice(lambd, mu, f, n):
    """ n simulation step of M|M|1, rates lambda, mu, initial filling f
    
    events: a list of n+1 tuples (t, f)
    Fillings at the beginning and after n events
    """
    print "Simulate Modified $M | M | 1$"
    t = 0.
    evnt = [(t, f)]
    for i in range(n):
        if f>0:
            typ, dt = newestEvent(lambd/(f+1), mu) # MODIFIED ARRIVAL RATE
        else:
            typ, dt = newestEvent(lambd, -1.)
        t = t + dt
        if typ == 'arrival':
            f = f + 1
        else:
            f = f - 1
        evnt.append((t, f))
    return evnt
   
    
def simulation_driver(lambd, mu, f, n):
    """ Pictures for M|M|1: rates lambda und mu, initial filling f, n steps
    """
    # make sure the rates are floats
    lambd = float(lambd)
    mu = float(mu)
    
    sim_results = simulate_modified_postoffice(lambd, mu, f, n)
    fig1 = pl.figure()
    fig2 = pl.figure()
    fig3 = pl.figure()
    fig4 = pl.figure()
    filling(sim_results, fig1)
    #cumulativeFilling(sim_results, fig2)
    mf = averageFilling(sim_results, fig3)
    s = lambd/mu
    print "workload %.3f" %(1 - np.exp(-s)) # MODIFIED WORKLOAD
    print 'average filling: %.3g' % mf
    if lambd < mu:
        r = lambd/mu
        print 'expected filling: %.3g' % (s)
    dweillingTimes(sim_results, lambd, mu, fig4)

    
```

```python
# small run
simulation_driver(8., 12., 10, 100)
```

```python
# large run
simulation_driver(8., 12., 10, 10000)
```

```python
# overload
simulation_driver(8., 4., 0, 10000)
```
