```python
class ode:
	def __init__(self,r1,r2,l1,l2,gg,dt=0.01,tend=10.0):
		self.gg = gg
		self.l = array([l1,l2])
		self.S = array([r1,r2]).T
		
		##compute ode coefficients
		self.A = dot(self.S,np.dot(np.diag([l1,l2]),np.linalg.inv(self.S)))
		self.b = -dot(self.A,gg)
		
		##time discretization
		self.dt=dt
		self.tend=tend
		
	def printSystem(self):
		print 'Koeffizientenmatrix:'
		print self.A
		print 'Inhomogenietaet:'
		print self.b
		
	def evaluate(self,x):
		return dot(self.A,x) + self.b
		
	def getTrajectoryExact(self,x_start):
		t=arange(0,self.tend,self.dt)
		x_t=[]
		c=linalg.solve(self.S,x_start-self.gg)
		for ti in t:
			x_t.append(self.gg+c[0]*exp(self.l[0]*ti)*self.S[:,0] \
				+ c[1]*exp(self.l[1]*ti)*self.S[:,1])
		x_t=array(x_t)
		return t,x_t
		
	
```

```python
class ode_plotter:

    def __init__(self,ax1=subplot(121),ax2=subplot(122)):
		self.ax1=ax1
		self.ax2=ax2
		self.ax1.set_aspect('equal')
		
    def plot_trajectory(self,f,x_start):
		t,x_t=f.getTrajectoryExact(x_start)
		self.ax1.plot(x_t[:,0],x_t[:,1],'b',lw=2)
		self.ax2.plot(t,x_t[:,0],'r',lw=2)
		self.ax2.plot(t,x_t[:,1],'g',lw=2)
		
		
    def plot_arrows(self,f,n=20):
		##calculating the box size of the arrow plot
		xlim=self.ax1.get_xlim()
		ylim=self.ax1.get_ylim()
		extend_x=max([f.gg[0]-xlim[0],-f.gg[0]+xlim[1]])
		extend_y=max([f.gg[1]-ylim[0],-f.gg[1]+ylim[1]])
		extend=max([extend_x,extend_y])
	
		##defining the starting points of the arrows
		X,Y = np.mgrid[f.gg[0]-extend:f.gg[0]+extend:n*1j,\
			f.gg[1]-extend:f.gg[1]+extend:n*1j]
		pos = np.array([X.flatten(),Y.flatten()])
	
	
		u=np.zeros(X.flatten().size)
		v=np.zeros(X.flatten().size)

		for i in range(X.flatten().size):
			result =f.evaluate(pos.T[i]) 
			u[i]=result[0]
			v[i]=result[1]

		self.ax1.quiver(X,Y,u,v,scale=60)
		self._plot_r(f.S[:,0],f.gg,extend)
		self._plot_r(f.S[:,1],f.gg,extend)
        
    def _plot_r(self,r,gg,extend):
		if abs(r[0])<abs(r[1]):
			x_start=np.array([gg[0]-r[0]/r[1]*extend,gg[0]-extend])
			x_end =np.array([gg[0]+r[0]/r[1]*extend,gg[0]+extend])
		else:
			x_start=np.array([gg[1]-extend,gg[1]-r[1]/r[0]*extend])
			x_end=np.array([gg[1]+extend,gg[1]+r[1]/r[0]*extend])
			
		self.ax1.plot([x_start[0],x_end[0]],[x_start[1],x_end[1]],'k')
```

```python
r1 = np.array([1.,0.0])
r2 = np.array([0.,1.])

l1 = -1
l2 = -2

##equilibrium
gg = np.array([1.0,1.0])

f_ode=ode(r1,r2,l1,l2,gg)
f_ode.printSystem()

fig=figure(figsize=(16,10))
plotter=ode_plotter(fig.add_subplot(121),fig.add_subplot(122))

plotter.plot_trajectory(f_ode,np.array([2.0,4.0]))
plotter.plot_arrows(f_ode)

```
